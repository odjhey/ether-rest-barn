```python
import json
modules_file_path = "./modules_data.jsonl"
modules_raw_str=open(modules_file_path, "r").read().splitlines()
modules_raw_str
modules_json=[]
modules_json = [json.loads(m) for m in modules_raw_str]
print(len(modules_json))
modules_json[:3]
  
```

```python
import pandas as pd
df=pd.read_json(modules_file_path, lines=True)
df.head()
```

```python
by_slot_name=df.groupby(["slot", "name"]).size().sort_values(ascending=False)
by_slot_name
```

```python
df['slot'].unique, df['name'].unique()
```

```python
#df[df['name'] == 'TO}']
df[df['name'] == '7.\nTO}']
```

```python
import matplotlib.pyplot as plt
slot_counts=df['slot'].value_counts()
plt.stairs(slot_counts.tolist())
```

```python
pXX=slot_counts.quantile(0.80)
slot_counts[slot_counts>pXX]
```

```python
cores=df[df['slot'].str.contains("Core")]
cores_size=cores.groupby(["slot", "main_stat"]).size().sort_values(ascending=False)
pXX=cores_size.quantile(0.90)
plt.stairs(cores_size[cores_size > pXX])
```

```python
pxx_cores_size=cores_size[cores_size>20]
pxx_cores_size.unstack().plot(kind="bar",rot=55)
```

```python
main_stat_noval=df["main_stat"].str.split("+", expand=True)[0]
df["main_stat_noval"] = main_stat_noval
df.groupby("main_stat_noval").size().plot(kind="bar")
```

```python
slot_main_grp=df.groupby(["main_stat_noval", "slot"]).size()
pxx=slot_main_grp.quantile(0.95)
slot_main_grp[slot_main_grp > pxx]
```
